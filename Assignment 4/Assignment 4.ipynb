```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv(r"C:\Users\kshit\Downloads\archive\netflix.csv")
df.head()
```

```python

df.head()
```

```python

df.isnull().sum()
```

```python

# Unique values in categorical columns
categorical_columns = ['gender', 'subscription_type', 'region', 'device', 'payment_method', 'favorite_genre']
for col in categorical_columns:
    print(f"Unique values in {col}: {df[col].unique()}")
```

```python

num_cols = ['age', 'watch_hours', 'monthly_fee', 'churned']
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python

for col in categorical_columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col)
    plt.title(f"Count plot of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python

for col in ['subscription_type', 'region', 'device']:
    df_grouped = df.groupby(col)[['watch_hours', 'monthly_fee']].mean().reset_index()
    df_grouped.plot(x=col, kind='bar', figsize=(8,5), title=f"Avg Watch Hours & Monthly Fee by {col}")
    plt.show()
```

```python

genre_avg_watch = df.groupby('favorite_genre')['avg_watch_time_per_day'].mean().sort_values(ascending=False)
genre_avg_watch.plot(kind='bar', figsize=(8, 4), title='Avg Watch Time per Day by Genre')
plt.ylabel("Avg Watch Time (hrs)")
plt.show()
```

```python

for col in ['gender', 'region', 'subscription_type', 'payment_method']:
    churn_rate = df.groupby(col)['churned'].mean().sort_values(ascending=False)
    churn_rate.plot(kind='bar', figsize=(7, 4), title=f"Churn Rate by {col}")
    plt.ylabel("Churn Rate")
    plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```


Here are some possible insights you can derive (complete with your actual findings):

1. Customers with lower `watch_hours` tend to churn more.
2. `Premium` users show a lower churn rate compared to `Basic` users.
3. The `North America` region has the highest average watch hours.
