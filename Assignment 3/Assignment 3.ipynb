```python
#PART 4: Students Performance
#1. Display average marks in math, reading, and writing.
#2. Who performed better in writing â€“ males or females?
#3. Add a new column for total score and percentage.
#4. Plot a bar chart showing average scores grouped by parental level of education.
#5. Find students who scored above 90% in all subjects'

students=pd.read_csv('StudentsPerformance.csv')
students
```

```python
#1. Display average marks in math, reading, and writing.
avg_math = students['math score'].mean()
avg_reading = students['reading score'].mean()
avg_writing = students['writing score'].mean()

print("Average Math Score:", avg_math)
print("Average Reading Score:", avg_reading)
print("Average Writing Score:", avg_writing)
```

```python
#2. Who performed better in writing â€“ males or females?
avg_writing_by_gender = students.groupby('gender')['writing score'].mean()
print("Average Writing Score by Gender:")
print(avg_writing_by_gender)
```

```python
#3. Add a new column for total score and percentage.
students['total score'] = students['math score'] + students['reading score'] + students['writing score']
students['percentage'] = students['total score'] / 3
students
```

```python
#4. Plot a bar chart showing average scores grouped by parental level of education.
avg_scores_by_parent = students.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
avg_scores_by_parent.plot(kind='bar', figsize=(10, 6))
plt.title('Average Scores by Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#5. Find students who scored above 90% in all subjects'
high_achievers = students[(students['math score'] > 90) & 
                          (students['reading score'] > 90) & 
                          (students['writing score'] > 90)]

print("\nStudents who scored above 90 in all subjects:")
print(high_achievers)
```
